```python
# 11.File Merger: Read content from file1.txt and file2.txt, then combine them into merged.txt.

with open("log.txt", "r") as f1:
    data1 = f1.read()

with open("bio.txt", "r") as f2:
    data2 = f2.read()

with open("merged.txt", "w") as merged:
    merged.write(data1)
    merged.write("\n")
    merged.write(data2)

print("Files merged successfully.")
```

```python
# 15.	Size Checker: Read a large file and print how many lines it has without loading the whole file into memory at once (use a for loop over the file object).

count = 0

with open("bio.txt", "r") as file:
    for line in file:
        count += 1

print("Total number of lines:", count)
```

```python
# 16.	Vocabulary Builder: Read a text file and create a dictionary where keys are words and values are how many times each word appears (Word Frequency).

with open("bio.txt", "r") as file:
    text = file.read().lower()

words = text.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
# 17.	Stopword Remover: Given a list of "stopwords" (the, is, at), read a file and remove these specific words from the text.


stopwords = ["the", "is", "at"]

with open("bio.txt", "r") as file:
    text = file.read()

words = text.split()

filtered_words = []

for word in words:
    if word.lower() not in stopwords:
        filtered_words.append(word)

result = " ".join(filtered_words)

print(result)
```

```python
#18.	Sentence Splitter: Read a long paragraph from a file and save each sentence into a new line in a different file.

with open("bio.txt", "r") as file:
    text = file.read()

sentences = text.split(".")

with open("log.txt", "w") as file:
    for sentence in sentences:
        sentence = sentence.strip()
        if sentence:
            file.write(sentence + ".\n")

print("Sentences saved successfully.")
```

```python
# 19.	Data Cleaning: Read a file and remove all numbers and special characters, leaving only alphabet letters.

with open("log.txt", "r") as file:
    text = file.read()

clean_text = ""

for ch in text:
    if ch.isalpha() or ch.isspace():
        clean_text += ch

print(clean_text)
```

```python
# 20.	Automation: Write a script that checks if a file exists; if it does, append "Updated" to it; if not, create it.

import os

filename = "bio.txt"

if os.path.exists(filename):
    with open(filename, "a") as file:
        file.write("\nUpdated")
    print("File updated.")
else:
    with open(filename, "w") as file:
        file.write("Updated")
    print("File created.")
```
