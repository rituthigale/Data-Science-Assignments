# Basic List Comprehensions 

```python
# Squares: Create a list of squares for numbers 1 to 10.
squares = [i**2 for i in range(1, 11)] 
squares
```

```python
# Upper Case: Given names = ['om', 'sai', 'ram'], convert them all to uppercase in one line.
upper_case = [name.upper() for name in ['om', 'sai', 'ram']]
upper_case
```

```python
# Filtered Range: Create a list of numbers from 1 to 50 that are divisible by 5.
divisible_by_5 = [num for num in range(1, 51) if num % 5 == 0]
divisible_by_5
```

```python
# String Lengths: Given a list of words, create a list containing the length of each word.
strings = [len(w) for w in ['avinash', 'avani', 'shreya']]
strings
```

```python
# First Letter: Extract the first character of every string in a list.
extract = [string[0] for string in ['avinash', 'avani', 'shreya']]
extract
```

# Conditionals in Comprehension

```python
# Even/Odd Labels: Create a list that says "Even" if the number is even and "Odd" if it is odd for range(1, 6).
# Hint: [ "Even" if i % 2 == 0 else "Odd" for i in range(1, 6) ]
even_odd = [ "Even" if i % 2 == 0 else "Odd" for i in range(1, 6) ]
even_odd
```

```python
# Positive Only: From data = [1, -5, 2, -8, 10], extract only the positive numbers.
positive_nums = [i for i in [1, -5, 2, -8, 10] if i > 0]
positive_nums
```

```python
# Vowel Remover: Given a string, create a list of all its characters except the vowels.
string = [char for char in "Hello, This is an example." if char.lower() not in 'aeiou']
string
```

```python
# Type Filter: Given l = [1, 'a', 2.5, 'b', 3], extract only the integers.
l = [1, 'a', 2.5, 'b', 3]
integers = [i for i in l if type(i) == int]
integers
```

```python
# Short Word Filter: From a list of fruits, keep only the ones with more than 5 letters.
fruits = ['Mango', 'Banana', 'Cherry', 'Pear']
[fruit for fruit in fruits if len(fruit) > 5]
```

#  Dictionary Comprehensions

```python
# Number Mapping: Create a dictionary where the key is a number (1-5) and the value is its cube.
{i : i**3 for i in range(1, 6)}
```

```python
# Name Lengths: Create a dictionary where keys are names and values are their lengths.
names = ['Riya', 'Piya', 'Niya']
{n : len(n) for n in names}
```

```python
# Price Conversion: Given prices_in_usd = {'apple': 1.0, 'bread': 2.5}, create a new dictionary with prices doubled.
prices = {'apple': 1.0, 'bread': 2.5}
{k : v * 2 for k,v in prices.items()}
```

```python
# Character Count: Create a dictionary counting occurrences of each character in "ALPHABET" (use .count()).
{k : "ALPHABET".count(k) for k in "ALPHABET"}
```

```python
# Conditional Dict: Create a dictionary of numbers 1-10, but only include those whose square is greater than 50.
{i : i**2 for i in range(1, 11) if i**2 > 50 }
```

```python
# Matrix Creation: Create a 3x3 matrix of zeros: [[0, 0, 0] for i in range(3)].
[[0 for _ in range(3)] for _ in range(3)]
```

```python
# Zip to Dict: Given keys = ['ID', 'Name'] and vals = [101, 'Python'], use zip() inside a comprehension to make a dictionary.
keys = ['ID', 'Name']
vals = [101, 'Python']
{k: v for k, v in zip(keys, vals)}
```

```python
# Dictionary Swap: Given d = {'a': 1, 'b': 2}, swap keys and values: {v: k for k, v in d.items()}.
{v: k for k, v in {'a': 1, 'b': 2}.items()}
```
